```python
import os
import sys
import torch
import numpy as np
from pygmid import Lookup as lk

import ComputingParameters
import EnviromentSetup
import TrainingLoop
import AgentSetup

print(f'PyTorch  : {torch.__version__}')
print(f'CUDA     : {torch.cuda.is_available()} — {torch.cuda.get_device_name(0) if torch.cuda.is_available() else "CPU only"}')
print(f'Python   : {sys.version.split()[0]}')
```

```python
# Lookup table paths 
NCH_FILE = './XT018/LVT/180nch.mat'   
PCH_FILE = './XT018/LVT/180pch.mat'

NCH = lk(NCH_FILE)
PCH = lk(PCH_FILE)
print(f'NCH loaded: {NCH_FILE}')
print(f'PCH loaded: {PCH_FILE}')

# ── Device lengths 
LN = 1.4   # um  — NMOS channel length
LP = 1   # um  — PMOS channel length

# Fixed circuit constants 
VDD    = 1.2     # V
CFB    = 1e-12   # F  — fixed feedback cap
CL     = 1e-12   # F  — load cap

amp = ComputingParameters.CurrentReuse(
    NCH, PCH,
    VDD    = VDD,
    LN     = LN,
    LP     = LP,
    Cfb    = CFB,
    CL     = CL,
)

print(f'\nAmp object ready')
print(f'  LN/LP = {LN}/{LP} um  (ratio = {LN/LP:.2f})')
```

```python
agent, log = TrainingLoop.train(
    PPOAgent        = AgentSetup.PPOAgent,
    EnviromentSetup = EnviromentSetup,
    amp_functions   = amp,
    output_dir      = 'runs/3rd_Run_1000_episodes',
)
```
